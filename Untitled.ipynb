```python
#Import libraries
import pandas as pd
import csv

import os
```

```python
#Read File

SchoolDataFile = "C:\\Users\\surbh\\Downloads\\Gurpreet\\Gurpreet DA Study\\Modules\\pandas-challenge\\Resources\\schools_complete.csv"
StudentDataFile = "C:\\Users\\surbh\\Downloads\\Gurpreet\\Gurpreet DA Study\\Modules\\pandas-challenge\\Resources\\students_complete.csv"

# Read both Data into Pandas DataFrames
SchoolData = pd.read_csv(SchoolDataFile)
StudentData = pd.read_csv(StudentDataFile)

# Merge the Data in Single DataFrame  
CompleteData = pd.merge(StudentData, SchoolData, how="left", on=["school_name", "school_name"])
CompleteData.head()
```

```python
StudentName = CompleteData["student_name"]
StudentName
```

```python
#Calculate Total # of School
SchoolCount =  len(CompleteData["school_name"].unique())
SchoolCount
```

```python
#Calculate the total number of students

StudentCount =CompleteData["Student ID"].count()
StudentCount
```

```python
#Calculate the total budget
TotalBudget = SchoolData["budget"].sum()
TotalBudget
```

```python
#Calculate the average math score
AverageMath = StudentData["math_score"].mean()
AverageMath
```

```python
#Calculate the average reading score
AverageReading = StudentData["reading_score"].mean()
AverageReading
```

```python
#% of Students passing Math Score (70 or greater)
PassingMathCount = CompleteData[(CompleteData["math_score"] >= 70)].count()["student_name"]
MathPercentage = PassingMathCount / float(StudentCount) * 100
MathPercentage
```

```python
#% of Students passing Read Score (70 or greater)
PassingReadingCount = CompleteData[(CompleteData["reading_score"] >= 70)].count()["student_name"]
ReadPercentage = PassingReadingCount / float(StudentCount) * 100
ReadPercentage
```

```python
#OverallPassing

PassingMathCount = CompleteData[
    (CompleteData["math_score"] >= 70) & (CompleteData["reading_score"] >= 70)].count()["student_name"]
OverallPassing = PassingMathCount /  float(StudentCount) * 100
OverallPassing
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
# Display the DataFrame

DistrictSummary = [{"Total Schools": SchoolCount, 
            "Total Students": StudentCount, 
            "Total Budget": TotalBudget, 
            "Average Math Score":  round(AverageMath,2), 
            "Average Reading Score":  round(AverageReading,2), 
           "% Passing Math": round(MathPercentage,2),
           "% Passing Reading": round(ReadPercentage,2),
            "% Overall Passing Rate": round(OverallPassing,2)}]


DistrictSummary
```

```python
CompleteData["Passing Math"] =CompleteData ["math_score"] >= 70
CompleteData["Passing Reading"] = CompleteData["reading_score"] >= 70

CompleteData
```

```python
#Create a dataframe and formatting

DistrictTable = pd.DataFrame(DistrictSummary)

DistrictTable["% Passing Math"] = DistrictTable["% Passing Math"].map("{:,.2f}%".format)
DistrictTable["Total Students"] = DistrictTable["Total Students"].map("{:,}".format)
DistrictTable["Total Budget"] = DistrictTable["Total Budget"].map("${:,.2f}".format)
DistrictTable["% Passing Reading"] = DistrictTable["% Passing Reading"].map("{:,.2f}%".format)


DistrictTable
```

```python
#Using Groupby

SchoolGroup = CompleteData.groupby(["school_name"]).mean()
SchoolGroup["Per Student Budget"] = SchoolGroup["budget"]/SchoolGroup["size"]
SchoolGroup["% Passing Math"] = round(SchoolGroup["math_score"]*100,2)
SchoolGroup["% Passing Reading"] = round(SchoolGroup["reading_score"]*100,2)
SchoolGroup["% Overall Passing Rate"] = round(((SchoolGroup["Passing Math"] + SchoolGroup["Passing Reading"])/2)*100,3)
SchoolGroup
```

```python
SchoolSummary = pd.merge(SchoolGroup, SchoolData, how="left", on=["school_name", "school_name"])
del SchoolSummary['size_y']
del SchoolSummary['budget_y']
del SchoolSummary['Student ID']
del SchoolSummary['School ID_x']
SchoolSummary
```

```python
SchoolSummary_df = pd.DataFrame({"School Name":  SchoolSummary["school_name"],
                                "School Type": SchoolSummary["type"],
                               "Total Students":SchoolSummary["size_x"],
                               "Total School Budget": SchoolSummary["budget_x"],
                               "Per Student Budget":SchoolSummary["Per Student Budget"], 
                               "Average Math Score":round(SchoolSummary["math_score"],2),
                               "Average Reading Score":round(SchoolSummary["reading_score"],2), 
                               "% Passing Math": SchoolSummary["% Passing Math"],
                               "% Passing Reading": SchoolSummary["% Passing Reading"],
                               "% Overall Passing Rate": SchoolSummary["% Overall Passing Rate"]}) 
SchoolSummary_df
```

```python
SchoolSummary_df["Total Students"] = SchoolSummary_df["Total Students"].map("{:,.0f}".format)
SchoolSummary_df["Total School Budget"] = SchoolSummary_df["Total School Budget"].map("${:,.2f}".format)
SchoolSummary_df["Per Student Budget"] = SchoolSummary_df["Per Student Budget"].map("${:,.2f}".format)

SchoolSummary_df
```

```python
#TopSchool Result
TopSchool = SchoolSummary_df.sort_values(["% Overall Passing Rate"], ascending=False)
TopSchool.head()
```

```python
#BottomSchool Result
BottomSchool = SchoolSummary_df.sort_values(["% Overall Passing Rate"], ascending=True)
BottomSchool.head()
```
